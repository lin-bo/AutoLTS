```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image

root = '/Users/bolin/Library/CloudStorage/OneDrive-UniversityofToronto/AutoLTS/'
dataset_root = '/Users/bolin/Library/CloudStorage/OneDrive-UniversityofToronto/Streetview2LTS/'
```

# check mis-classified images

```python
def get_mis_img_idx(filename, true=0, pred=0):
    res = pd.read_csv(root + f'res/prediction/{filename}.csv')
    res['idx'] = res['idx'].astype(int)
    if true >= 1:
        res = res[res['true'] == true]
    if pred >= 1:
        res = res[res['pred'] == pred]
    return res['idx'].values
```

```python
filename = 'Res50FC_8520206_4_validation'
true = 3
pred = 4
idx = get_mis_img_idx(filename, true=true, pred=pred)
```

```python
def show_img(filename, pred=0, true=0, k=10):
    idx = get_mis_img_idx(filename, true=true, pred=pred)
    np.random.shuffle(idx)
    for i in idx[:k]:
        img = Image.open(dataset_root + f'dataset/{i}.jpg')
        fig = plt.figure()
        plt.imshow(img)
        plt.xticks([])
        plt.yticks([])
        plt.title(f'True: {true}, Pred: {pred}')
        plt.show()
```

```python
filename = 'Res50FC_8520206_4_validation'
true = 3
pred = 4
show_img(filename, true=3, pred=4, k=3)
```
